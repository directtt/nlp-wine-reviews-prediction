Na pierwszy rzut oka możemy wyszczególnić następujące rodzaje cech oraz ich opisy: <br><br>
Zmienne jakościowe:
- `country` - The country that the wine is from.

- `region_1`- The wine growing area in a province or state (ie Napa).
- `region_2`- Sometimes there are more specific regions specified within a wine growing area (ie Rutherford inside the Napa Valley), but this value can sometimes be blank.
- `designation` - The vineyard within the winery where the grapes that made the wine are from.
- `winery` - The winery that made the wine.
- `variety` - The type of grapes used to make the wine (ie Pinot Noir).<br><br>

Zmienna tekstowa:
- `description`<br><br>

Zmienne numeryczne:
- `price` - The cost for a bottle of the wine.
- `points` - The number of points WineEnthusiast rated the wine on a scale of 1-100 (though they say they only post reviews for wines that score &gt;=80).

### Missing values

```python
def missing(df):
    perc_missing = (df.isnull().sum() * 100 / len(df)).round(2)
    missing_value_df = pd.DataFrame({'num_missing': df.isnull().sum(),
                                     'perc_missing': perc_missing})
    missing_value_df = missing_value_df[missing_value_df.num_missing != 0]
    return missing_value_df.sort_values(by='perc_missing', ascending=False)
```

### Target variable distribution

```python
def show_dist(df, var):
    fig = plt.figure(constrained_layout=True, figsize=(10, 5))
    grid = gridspec.GridSpec(ncols=3, nrows=1, figure=fig)
    
    ax1 = fig.add_subplot(grid[0, :2])
    ax1.set_title('Histogram of %s' % var)
    sns.distplot(df.loc[:, var], ax=ax1);
    
    ax2 = fig.add_subplot(grid[0, 2])
    ax2.set_title('Box Plot of %s' % var)
    sns.boxplot(y=df.loc[:, var], ax=ax2);
```

### Punctuations Removal

```python
import string
import re

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_punctuation_extra(text):
    text = re.sub('[—-]', ' ', text)
    text = re.sub('[”“]', '', text)
    return text

df['description'] = df['description'].apply(remove_punctuation_extra)
df['description'] = df['description'].apply(remove_punctuation)
```

```python
df['description'].head()
```

### Stopwords removal & tokenization

```python
df['description'].str.split(' ').str.len().sum() # 6kk words
```

```python
import nltk
nltk.download('stopwords')

def remove_stopwords(text):
    stopwords = nltk.corpus.stopwords.words('english')
    removed = [i for i in text.split() if i not in stopwords]
    return removed
```

```python
df['description']= df['description'].apply(remove_stopwords)
df['description'].head()
```

### Stemming

```python
from nltk.stem.porter import PorterStemmer

stemmer = PorterStemmer()
def stemming(text):
    stem_text = [stemmer.stem(word) for word in text]
    return stem_text
```

```python
df_wo_stemming = df.copy()

df['description']=df['description'].apply(stemming)
df['description'].head()
```

Dobrym pomysłem wydaje się być zastosowanie CatBoosta, chociażby w celu wykorzystania jego encodingu dla naszych mocno zróżnicowanych cech kategorycznych oraz przejrzenia, które cechy poza wcześniej wykorzystanym `description` mają wpływ na zmienną wyjściową.

```python
cb = CatBoostClassifier(loss_function='MultiClass',
                        task_type='CPU',
                        learning_rate=0.45,
                        iterations=2500,
                        random_seed=42)
```

```python
X = df_catboost.drop(columns='points')
y = df_catboost['points']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
```

```python
train_pool = Pool(
    data=X_train,
    label=y_train,
    text_features=['description'],
    cat_features=['country', 'designation', 'province', 'region_1', 'variety', 'winery']
)

valid_pool = Pool(
    data=X_test, 
    label=y_test,
    text_features=['description'],
    cat_features=['country', 'designation', 'province', 'region_1', 'variety', 'winery']
)
```

```python
cb.fit(
    train_pool,
    eval_set=valid_pool,
    verbose=100,
    plot=True
);
```

```python
feature_importances = cb.get_feature_importance()
feature_names = X_train.columns
for score, name in sorted(zip(feature_importances, feature_names), reverse=True):
    print('{}: {}'.format(name, score))
```

## RandomForest (Text + Price)

Skoro MI pokazywał, że tylko zmienna `price` jest skorelowana ze zmienną wyjściową, użyjmy tylko jej + `description` do trenowania modelu.

```python
from sklearn.model_selection import train_test_split

X = df_other_enc[['description', 'price']]
y = df_other_enc['points']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2, stratify=y)
```

```python
X_train_desc = X_train.description.copy()
X_test_desc = X_test.description.copy()
X_train = X_train.drop(columns='description').reset_index(drop=True)
X_test = X_test.drop(columns='description').reset_index(drop=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=1000)
X_train_tfidf = pd.DataFrame(tfidf.fit_transform(X_train_desc).toarray())
X_test_tfidf = pd.DataFrame(tfidf.transform(X_test_desc).toarray())
```

```python
X_train = pd.concat([X_train, X_train_tfidf], axis=1)
X_test = pd.concat([X_test, X_test_tfidf], axis=1)
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)

print(classification_report(y_test, rf.predict(X_test)))
```
